# Tidiness

## Issue #1

### Define
- create a column for image prediction and confidence level

### Code

```python
dog_type = []
confidence_list = []

def image(image_prediction_cleaning):
    if image_prediction_cleaning['p1_dog'] == True:
        dog_type.append(image_prediction_cleaning['p1'])
        confidence_list.append(image_prediction_cleaning['p1_conf'])
    elif image_prediction_cleaning['p2_dog'] == True:
        dog_type.append(image_prediction_cleaning['p2'])
        confidence_list.append(image_prediction_cleaning['p2_conf'])
    elif image_prediction_cleaning['p3_dog'] == True:
        dog_type.append(image_prediction_cleaning['p3'])
        confidence_list.append(image_prediction_cleaning['p3_conf'])
    else:
        dog_type.append('Error')
        confidence_list.append('Error')

#series objects having index the image_prediction_clean column.        
image_prediction_cleaning.apply(image, axis=1)

#create new columns
image_prediction_cleaning['dog_type'] = dog_type
image_prediction_cleaning['confidence_list'] = confidence_list
```

```python
#drop rows that has prediction_list 'error'
image_prediction_cleaning = image_prediction_cleaning[image_prediction_cleaning['dog_type'] != 'Error']

#drop rows not used in analysis
image_prediction_cleaning = image_prediction_cleaning.drop(['img_num', 'p1', 
                                                      'p1_conf', 'p1_dog', 
                                                      'p2', 'p2_conf', 
                                                      'p2_dog', 'p3', 
                                                      'p3_conf', 
                                                      'p3_dog'], 1)
```

```python
twi_arch_cleaning = pd.melt(twi_arch_cleaning, id_vars=['tweet_id',                                          
                                                                'timestamp',
                                                                'text',
                                                                'rating_over_10',
                                                                'name'],
                               var_name='dogs', value_name='dogs_stage')
#drop dogs
twi_arch_cleaning = twi_arch_cleaning.drop('dogs', 1)

#sort
twi_arch_cleaning = twi_arch_cleaning.sort_values('dogs_stage').drop_duplicates(subset='tweet_id', 
                                                                                        keep='last')
                            
```
